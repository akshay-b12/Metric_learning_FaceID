```python
from pylmnn import LargeMarginNearestNeighbor as LMNN
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_olivetti_faces
from sklearn.decomposition import PCA
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import cv2
import dlib
from imutils import face_utils
from skimage import data, feature, color, filters, img_as_float, io
from matplotlib import pyplot as plt
import math
```

```python
class Cldml:
    def __init__(self, face_detect_path, shape_pred_path):
        ## Initialize OpenCV's face detector
        self.face_cascade = cv2.CascadeClassifier(face_detect_path)

        ## Initialize DLib's shape predictor
        self.shape_predictor = dlib.get_frontal_face_detector()
        self.predictor = dlib.shape_predictor(shape_pred_path)

        ## Initilize OpenCV's SIFT detector
        self.sift_detector = cv2.xfeatures2d.SIFT_create()

        ## Data path
        #self.data_path = datapath

        ## Initialize LDML instance
        #ldml_instance = LDML()

    def get_face(self,image):
        if len(image.shape) > 2:
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        else:
            gray = image
        faces = self.face_cascade.detectMultiScale(gray, 1.3, 5)
        if faces is None:
            return faces, False
        
        img = image.copy()
        for (x,y,w,h) in faces:
            img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),5)
        
        plt.figure(figsize=(10,10))
        plt.axis("off")
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.show()
        return faces, True
    
    def get_landmark_points(self, image, rect):
        if len(image.shape) > 2:
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        else:
            gray = image
        rect = dlib.rectangle(rect[0], rect[1], rect[0]+rect[2], rect[1]+rect[3])
        shape = self.predictor(gray, rect)
        shape = face_utils.shape_to_np(shape)
        #print(shape)
        #kp_idx = [36, 39, 42, 45, 31, 33, 35, 48, 54]
        #for i in kp_idx:
		    #    cv2.circle(image, (shape[i][0], shape[i][1]), 10, (0, 0, 255), -1)
        #plt.figure(figsize=(10,10))
        #plt.axis("off")
        #plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        #plt.show()
        return shape

    def get_sift_descriptors(self,images):
        kp_idx = [36, 39, 42, 45, 31, 33, 35, 48, 54]
        scale = 0.7
        descriptors = []
        for i in range(len(images)):
            for img in images[i]:
                img_des = []
                if len(img.shape) > 2:
                    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                else:
                    gray = img
                gray = gray.astype(np.uint8)
                face_rect = (0,0,gray.shape[1]-1, gray.shape[0]-1)
                landmarks = self.get_landmark_points(gray, face_rect)
                keypoints = [cv2.KeyPoint(landmarks[i][0], landmarks[i][1], 1) for i in kp_idx]
                #print(keypoints[0].pt[1])
                for j in range(3):
                    kp,des = self.sift_detector.compute(gray,keypoints)
                    des = des.flatten()
                    img_des.append(des)
                    if j == 2:
                        break
                    gray = cv2.resize(gray,None,fx=0.7,fy=0.7, interpolation=cv2.INTER_AREA)
                    keypoints = [cv2.KeyPoint(keypoints[k].pt[0]*0.7, keypoints[k].pt[1]*0.7, 1) for k in range(len(keypoints))]
                img_des = np.array(img_des).flatten()
                descriptors.append(img_des)
                #img=cv2.drawKeypoints(gray,kp,gray,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
                #plt.figure(figsize=(10,10))
                #plt.axis("off")
                #plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
                #plt.show()
                #cv2.imwrite('sift_keypoints.jpg',img)
            print("Pair ", i+1, " computed")
        return descriptors

    def create_label(self,path):
        file = open(path, mode = 'r')
        label_count = 0
        label_dict = {}
        labels = []
        line = file.readline()
        txt = line.split()
        sets = int(txt[0])
        count = int(txt[1])
        for i in range(sets):
            for j in range(count):
                line = file.readline()
                txt = line.split()
                if txt[0] not in label_dict:
                    label_dict[txt[0]] = label_count+1
                    label_count += 1
                    labels.append(label_count)
                    labels.append(label_count)
                else:
                    labels.append(label_dict[txt[0]])
                    labels.append(label_dict[txt[0]])
            for j in range(count):
                line = file.readline()
                txt = line.split()
                if txt[0] not in label_dict:
                    label_dict[txt[0]] = label_count+1
                    label_count += 1
                    labels.append(label_count)
                else:
                    labels.append(label_dict[txt[0]])
                if txt[2] not in label_dict:
                    label_dict[txt[2]] = label_count+1
                    label_count += 1
                    labels.append(label_count)
                else:
                    labels.append(label_dict[txt[2]])
        return labels
```

```python
from sklearn.datasets import fetch_lfw_pairs
lfw_pairs_train = fetch_lfw_pairs(subset='10_folds', funneled=True, resize = 1, color=False)
```

```python
face_path = "/content/drive/My Drive/computer_vision/CV_project/haarcascade_frontalface_default.xml"
shape_path = "/content/drive/My Drive/computer_vision/CV_project/shape_predictor_68_face_landmarks.dat"
ldml_obj = Cldml(face_path, shape_path)
pairs_path = "/content/drive/My Drive/computer_vision/CV_project/pairs.txt"
labels = ldml_obj.create_label(pairs_path)
print(len(labels),max(labels))
des = ldml_obj.get_sift_descriptors(lfw_pairs_train.pairs)
```

```python
pca = PCA(n_components=200)
output_pca = pca.fit_transform(des)
print(output_pca.shape)
```

```python
X_train = output_pca[0:11400]
X_test = output_pca[11400: ]
y_train = labels[0:11400]
y_test = labels[11400:]
```

```python
k_train, k_test, n_components, max_iter = 3, 3, output_pca.shape[1], 180

# Instantiate the metric learner
lmnn = LMNN(n_neighbors=k_train, max_iter=max_iter, n_components=n_components)

# Train the metric learner
lmnn.fit(X_train, y_train)

# Fit the nearest neighbors classifier
knn = KNeighborsClassifier(n_neighbors=k_test)
knn.fit(lmnn.transform(X_train), y_train)

# Compute the k-nearest neighbor test accuracy after applying the learned transformation
lmnn_acc = knn.score(lmnn.transform(X_test), y_test)
print('LMNN accuracy on test set of {} points: {:.4f}'.format(len(X_test[0]), lmnn_acc))
```
